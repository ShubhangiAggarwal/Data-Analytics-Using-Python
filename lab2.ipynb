```python
#reading a .csv file
import pandas as pd 
data = pd.read_csv("KCLT.csv") 
# Preview the first 5 lines of the loaded data using data.head()
data
```

```python
#Find mean, median, mode, min and max for all numeric attributes
data['actual_mean_temp'].mean()  #calculate for only one attribute
x=data.mean(axis=0) #'0' for columns and '1' for rows
y=data.median(axis=0)
z=data.mode(axis=0)
a=data.min(axis=0)
b=data.max(axis=0)
print(x)
print(y)
print(z)
print(a)
print(b)
```

```python
#Plot min and max for actual_mean_temp, actual_min_temp and actual_max_temp using bar graph. Use blue for min and red for max.
import numpy as np
import matplotlib.pyplot as plt
# data to plot
n_groups = 3
means_max = (data['actual_mean_temp'].max(), data['actual_min_temp'].max(), data['actual_max_temp'].max())
means_min= (data['actual_mean_temp'].min(), data['actual_min_temp'].min(), data['actual_max_temp'].min())
# create plot
index = np.arange(n_groups) #numpy.arange([start, ]stop, [step, ]dtype=None)¶
#Return evenly spaced values within a given interval.
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, means_max , bar_width,
                 alpha=opacity,
                 color='red',
                 label='max')
 
rects2 = plt.bar(index + bar_width, means_min, bar_width,
                 alpha=opacity,
                 color='b',
                 label='min')

plt.xlabel('Attributes')
plt.ylabel('Min and Max')
plt.title('Analysis of Minimum and Maximum temperatures')
plt.show()
```

```python
#Plot two separate histograms for actual_mean_temp for months of Aug 2014 and March 2015
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1
data1=data.iloc[31:62, 0:2]
plt.hist(data1['actual_mean_temp'] , color='y')
plt.xlabel('Mean Temperature')
plt.ylabel('Frequency')
plt.title("Histogram for actual_mean_temp for Aug 2014")
plt.show()
data1=data.iloc[245:275, 0:2]
plt1.hist(data1['actual_mean_temp'] , color='r')
plt1.xlabel('Mean Temperature')
plt1.ylabel('Frequency')
plt1.title("Histogram for actual_mean_temp for March 2015")
plt1.show()
#Key observations:
#in Aug 2014, the actual mean temperature's mode is 78 where as  in march'15 its's 55 and 65.
#the values of mean temperature are varying less in Aug2014 then in March'15
#Aug2014 has much higher values than march 2015. Thus, having hotter weather.
```

```python
#correlation between actual_mean_temp and actual_min_temp and actual_mean_temp and actual_max_temp
import pandas as pd
df = pd.read_csv("KCLT.csv")
#correlation between actual_mean_temp and actual_min_temp
x=df['actual_mean_temp'].corr(df['actual_min_temp'])
#correlation between actual_mean_temp and actual_max_temp
y=df['actual_mean_temp'].corr(df['actual_max_temp'])
print(x)
print(y)
#keyobservations:
# actual_mean_temp is strongly corelated with actual_min_temp and actual_max_temp
# The degree of correlation is almost equal in both cases
```

```python
# Plot two scatter plots between following pairs of attributes: actual_mean_temp & actual_min_temp and 
#   actual_mean_temp & actual_max_temp

import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("KCLT.csv")

plt.scatter(data['actual_mean_temp'], data['actual_min_temp'],c='r')
plt.xlabel('actual_mean_temp')
plt.ylabel('actual_min_temp')
plt.show()

plt.scatter(data['actual_mean_temp'], data['actual_max_temp'],c='b') 
plt.xlabel('actual_mean_temp')
plt.ylabel('actual_max_temp')
plt.show()

#Key observation 
#with the increase in mean temperature, actual minimum and maximum temperature also increase.
```

```python
#make correlation matrix plot and scatter matrix plot taking all temperature attributes in dataset.

# Correction Matrix Plot

import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(r"KCLT.csv")
import numpy

temp_cols = data._get_numeric_data()
corr = temp_cols.corr() #correlation matrix
names =list(temp_cols) #get column names

# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,12,1)
plt.xticks(rotation='vertical')
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names,fontsize=8)
ax.set_yticklabels(names)
plt.show()



```
